```python
# https://et-poc-openai.openai.azure.com/openai/deployments/gpt-4o-2024-11-20/chat/completions?api-version=2025-01-01-preview
# 7fb0f62ba5bd43c1b427e1ecb6220af2

endpoint = os.getenv("ENDPOINT_URL", "https://et-poc-openai.openai.azure.com/")
deployment = os.getenv("DEPLOYMENT_NAME", "gpt-4o-2024-11-20")
subscription_key = os.getenv("AZURE_OPENAI_API_KEY", "7fb0f62ba5bd43c1b427e1ecb6220af2")

# Initialize Azure OpenAI Service client with key-based authentication
client = AzureOpenAI(
    azure_endpoint=endpoint, api_key=subscription_key, api_version="2025-01-01-preview",
)

# IMAGE_PATH = "YOUR_IMAGE_PATH"
# encoded_image = base64.b64encode(open(IMAGE_PATH, 'rb').read()).decode('ascii')
```

```python
#Prepare the chat prompt
chat_prompt = [
    {
        "role": "system",
        "content": [
            {
                "type": "text",
                "text": "You are an AI assistant that helps people find information."
            }
        ]
    }
]
```

```python
def llm_chat(type, msg):
    add_msg = {
        "role": type,
        "content": [
            {
                "type": "text", "text": msg
            }
        ]
    }

    chat_prompt.append(add_msg)
    
    # Include speech result if speech is enabled
    messages = chat_prompt

    # Generate the completion
    completion = client.chat.completions.create(
        model=deployment,
        messages=messages,
        max_tokens=16384,
        temperature=0.7,
        top_p=0.95,
        frequency_penalty=0,
        presence_penalty=0,
        stop=None,
        stream=False
    )

    add_msg = {
        "role": "system",
        "content": [
            {
                "type": "text", "text": completion.choices[0].message.content
            }
        ]
    }

    chat_prompt.append(add_msg)

    return completion.choices[0].message.content
```

```python
usr_prompt = '''
Share with me a final well drafter version, end to end, don't skip anything'''
```

```python
printmd(llm_chat('user', usr_prompt))
```

```python
from datetime import datetime

# Function to create Markdown content
def generate_markdown(chat_data):
    markdown_content = "## Chat Log\n\n"
    for entry in chat_data:
        role = entry['role'].capitalize()
        color_map = {
            "User": "cyan",
            "System": "green"
        }
        role_color = color_map.get(role, "black")
        # markdown_content += f"## {role}\n"
        markdown_content += f"<h2 style='color:{role_color};'>##{role}##</h2>\n"
        for content in entry['content']:
            if content['type'] == 'text':
                markdown_content += f"{content['text']}\n\n"
    return markdown_content

# Get current date and time for the filename
current_datetime = datetime.now().strftime("%Y%m%d_%H%M%S")
output_file = f"ChatLog_{current_datetime}.md"

# Generate Markdown content
markdown_content = generate_markdown(chat_prompt)

# Save to .md file
with open(output_file, "w") as file:
    file.write(markdown_content)

print(f"Markdown file './{output_file}'")
```
